#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url2 = 'https://www.imdb.com/list/ls009796553/'
````

````python
response = requests.get(url2)
response
````

````python
soup = BeautifulSoup(response.content)
````

````python
movie_list = soup.find_all('div', class_='lister-item mode-detail')
movie_list
````

````python
movies_list = []

for movie in soup.find_all('div', class_='lister-item mode-detail'):
    title = movie.find('h3').find('a').text
    year = movie.find('span', class_='lister-item-year').text.strip('()')
    summary = movie.find('p', class_=' ')

    movies_list.append(
        {
            'Title': title,
            'Year': year,
            'Summary': summary
        }
    )
````

````python
movies_list
````

````python
movies_df = pd.DataFrame(movies_list)
movies_df.head(10)
````
